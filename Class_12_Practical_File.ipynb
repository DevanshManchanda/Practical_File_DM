##**Q1 Write a menu driven program to do the following tasks with the help of user-defined functions**
**1) Find factorial of a number  
2) Check if the number is prime number or not   
3) Fibonacci series upto n terms  
4) To calculate  & return the sum of the series  
x/1! + x3/3! + x5/5! + x7/7! + …nterms**




```python
def fact(x):
    a=1
    for i in range (1,x+1):
        a*=i
    return a

def isprime(n):
    if n == 0 or n==1:
        return "Neither Prime nor Composite"
    if n < 0:
        n = 0-n
    start = 2
    end = (n//2) + 1
        
    for i in range(start, end):
        if n%i == 0:
            return 'Number is not prime'
    return 'Number is prime'

def fibo(n):
    f = 0
    s = 1
    print('Fibonacci Series:', end = ' ')
    print(f,s,end = ' ')
    for i in range(n-2):
        t = f+s
        print(t, end = ' ')
        f = s
        s = t 
        t = f+s 
    print()

def series(x,n):
    sum = 0
    for i in range(1,2*n,2):
        sum += (x**i)/fact(i)
    return sum


print(" ")
print(" ")
print("----------------------Output-------------------------")
print(" ")
print(" ")

print("MENU OPTIONS")
print("1. Find factorial of a number")
print("2. Check if the number is prime number or not ")
print("3. Fibonacci series upto n terms")
print('''4. To calculate & return the sum of the series
                      x/1! + x3/3! + x5/5! + x7/7! + …nterms''')
print("5. Exit")

while True:
    ch=int(input("Enter your Choice: "))
    if ch==1:
        num=int(input("Enter Number: "))
        print(fact(num))
    
    elif ch == 2:
        num = int(input("Enter Number: "))
        print(isprime(num))

    elif ch == 3:
        num = int(input("Enter Number: "))
        fibo(num)

    elif ch == 4:
        num = int(input("Enter n: "))
        var = int(input("Enter x: "))
        print(series(var, num))

    elif ch == 5:
        print('Thanks for using!')
        break
    
    else:
      print("Wrong Input")
```

###**Q7 Write a menu driven program to do the following tasks with the help of user-defined functions on Notes.txt**
**1)Replace 's' with 'S' in new file “Capital.txt”.**  
**2) Display & Count no. of words beginning with 'E' OR 'A'.**  
**3)Display the no. of lines beginning with 'a' or 'A' and ending with 'e' or   'E'.**  
**4)Replace word 'DELHI' with 'NEWDELHI' in new text file City.txt.**                                                                  

```python
def createfile():
    f1=open("Notes.txt","w")
    print("Enter text to the text file (Enter blank to stop)")
    while True:
        s=input()
        if len(s) ==0:
            break
        f1.write(s+"\n")
    f1.close()

def replace_S():
    f1=open("Notes.txt","r")
    f2=open("Capital.txt","w")
    s1=f1.read()
    for ch in s1:
        if ch=='s':
            f2.write('S')
        else:
            f2.write(ch)

    f1=open("Notes.txt","r")
    s=f1.read()
    print("Old Data")
    print(s)

    f2=open("Capital.txt","r")
    print("New Data")
    s2=f2.read()
    print(s2)


def count_ea():
    f=open("Notes.txt","r")
    s=f.read()
    a=s.strip()
    b=a.split()
    c=0

    for i in b:
        if i[0]=="A" or i[0]=="E":
            print(i)
            c+=1
    print(c)

def line_ea():
    f=open("Notes.txt","r")
    s=f.readlines()
    c=0

    for i in s:
        if i[0] in "Aa" or i[-2] in "Ee":
          print(i,end="")
          c+=1
    print(c)


def change_delhi():
    f1=open("Notes.txt","r")
    f2=open("City.txt","w")
    s1=f1.read()
    word_lst=s1.split()

    for word in word_lst:
        word=word.strip()
        if word.upper()=="DELHI":
            print(word)
            f2.write("NEWDELHI"+" ")
        else:
            f2.write(word+" ")
            
    f1=open("Notes.txt","r")
    s=f1.read()
    print("Old Data")
    print(s)

    f2=open("City.txt","r")
    print("New Data")
    s2=f2.read()
    print(s2)

print(" ")
print(" ")
print("----------------------Output-------------------------")
print(" ")
print(" ")

print("Creating File")
createfile()

print("MENU OPTIONS")
print("1. Replace 's' with 'S' in new file “Capital.txt”.")
print("2. Display & Count no. of words beginning with 'E' OR 'A'.")
print("3. Display the no. of lines beginning with 'a' or 'A' and ending with 'e' or 'E'.")
print("4. Replace word 'DELHI' with 'NEWDELHI' in new text file City.txt.")
print("5. Exit")


while True:
    ch=int(input("Enter your choice:  "))
    if ch==1:
        replace_S()
    elif ch==2:
        count_ea()
    elif ch==3:
        line_ea()
    elif ch==4:
        change_delhi()
    elif ch==5:
        print("End")
        break
    else:
        print("Wrong Input")
```

###**Q8 Create a binary file “STUDENT.DAT”, containing records of the following type:**
**(S_Admno, S_Name, Percentage)**  
**Where these three values are:**  
**S_Admno 	–	 Admission Number of student (string)**   
**S_Name	 –	 Name of student (string)**  
**Percentage	 –	 Marks percentage of student (float)** 

**Write a function in PYTHON that would read contents of the file “STUDENT.DAT” and display the details of those students whose percentage is above 75.**


```python
from pickle import *
def create_file():
    st_lst=[]

    ans="y"
    while ans=="y" or ans=="Y":
        S_Admno=str(input("Enter your Admission Number  "))
        S_Name=str(input("Enter your Name  "))
        Percentage=float(input("Enter your percentage "))
        st_lst.append([S_Admno,S_Name,Percentage])

        ans=input("Do you want to continue y/n: ")
    
    with open("Student.dat", "wb") as f:
        for rec in st_lst:
            print("Writing record :   ",rec)
            dump(rec,f)

def display_file():

    lst1=[]
    lst75=[]
    with open("Student.dat","rb") as f:
        try:
            while True:
                lst=load(f)
                lst1.append(lst)
        
        except EOFError:
            print()
    l=len(lst1)

    for i in range(l):
      if lst1[i][2]>=75:
        lst75.append(lst1[i])
    print("Students who have scored more than 75%")
    print(lst75) 

print(" ")
print(" ")
print("----------------------Output-------------------------")
print(" ")
print(" ")

create_file()
display_file()
```

###**Q9 Create a binary file “PRODUCT.DAT”. Following is the structure of each record in a data file named ”PRODUCT.DAT”    .{"prod_code":value,"prod_desc":value, "stock":value}** 

**The values for prod_code and prod_desc  are strings, and the value for stock is an integer.**   

**Write a function in PYTHON to update the file with a new value of stock.**   
 **The stock and the product_code of the product, whose stock is to be updated, are to be input during the execution of the function.**


```python
from pickle import *
def create_file():
    print('Creating file')
    st_lst=[]

    ans="y"
    while ans=="y" or ans=="Y":
        Prod_code=str(input("Enter product code  "))
        Prod_desc=str(input("Enter product description ")) 
        Stock=int(input("Enter product stock "))
        dict={"prod_code":Prod_code,"prod_desc":Prod_desc,"stock":Stock}
        st_lst.append(dict)

        ans=input("Do you want to continue y/n: ")
    
    with open("Product.dat", "wb") as f:
        for rec in st_lst:
            print("Writing record :   ",rec)
            dump(rec,f)


def change_stock():
    prod_code=str(input("Enter product code to change stock  "))
    new_stock=int(input("Enter product new stock "))

    lst1=[]
    lst_=[]
    with open("Product.dat","rb") as f:
        try:
            while True:
                lst=load(f)
                lst1.append(lst)
        except EOFError:
            print()

    l=len(lst1)

    for i in range (l):

        if lst1[i]['prod_code']==prod_code:
            lst1[i]['stock']=new_stock


    with open("Product.dat", "wb") as f:
        for rec in lst1:
            dump(rec,f)



def display_file():
    print("Displaying File Data")
    lst1=[]
    with open("Product.dat","rb") as f:
        try:
            while True:
                lst=load(f)
                lst1.append(lst)
        except EOFError:
            print()
    l=len(lst1)
    for i in range (l):
      print(lst1[i])

print(" ")
print(" ")
print("----------------------Output-------------------------")
print(" ")
print(" ")

create_file()
display_file()
change_stock()
display_file()
```

###**Q10 Write a menu-driven program for a file of employees [empno,ename,sal,deptt,City] using Functions for:-**
* **Create file**
* **Display all records**
* **Append a new employee**
* **Search for an employee whose empno is given**
* **Search & display for all occurances of given ename**
* **Sort emp records acc to their empno**
* **Delete an employee whose empno is given**
* **Update the sal of an employee whose empno is given**



```python
from pickle import *

def create_file():
    emp_lst=[]

    ans="y"
    while ans=="y" or ans=="Y":
        empno=int(input("Enter Employee's Number  "))
        ename=input("Enter Employee's Name  ")
        sal=float(input("Enter Employee's Salary  "))
        deptt=input("Enter Employee's Department  ")
        city=input("Enter Employee's City  ")

        emp_lst.append([empno,ename,sal,deptt,city])

        ans=input("Do you want to continue y/n: ")
    
    with open("Employee.dat", "wb") as d:
        for rec in emp_lst:
            print("Writing record :   ",rec)
            dump(rec,d)

def display_file():
    lst_1=[]

    print("Printing all records in file : ")
    with open("Employee.dat","rb") as d:
        try:
            while True:
                lst_=load(d)
                lst_1.append(lst_)
        
        except EOFError:
            print()
    l=len(lst_1)

    for i in range(l):
        print(lst_1[i])


def append_file():
    emp_lst=[]

    empno=int(input("Enter Employee's Number  "))
    ename=input("Enter Employee's Name  ")
    sal=float(input("Enter Employee's Salary  "))
    deptt=input("Enter Employee's Department  ")
    city=input("Enter Employee's City  ")
    emp_lst.append([empno,ename,sal,deptt,city])

    with open("Employee.dat", "ab+") as d:
        for rec in emp_lst:
            print("Writing record :   ",rec)
            dump(rec,d)

def find_empno():
    ep=int(input("Enter Employee's Number to search  "))

    lst1=[]
    with open("Employee.dat","rb") as d:
        try:
            while True:
                lst=load(d)
                lst1.append(lst)
        except EOFError:
            print()
    l=len(lst1)

    for i in range(l):
        if lst1[i][0]==ep:
            print(lst1[i])

def find_ename():
    en=input("Enter Employee's Name to search  ")

    lst1=[]

    with open("Employee.dat","rb") as d:
        try:
            while True:
                lst=load(d)
                lst1.append(lst)
        except EOFError:
            print()
    l=len(lst1)

    for i in range(l):
      if lst1[i][1]==en:
        print(lst1[i])

def sort_empno():
    lst1=[]

    with open("Employee.dat","rb") as d:
        try:
            while True:
                lst=load(d)
                lst1.append(lst)
        except EOFError:
            print()

        lst1.sort(key = lambda x: x[0])
        l=len(lst1)
        for i in range(l):
          print(lst1[i])

def del_empno():
    en=int(input("Enter Employee's Number to delete  "))

    lst1=[]

    with open("Employee.dat","rb") as d:
        try:
            while True:
                lst=load(d)
                lst1.append(lst)
        except EOFError:
            print()
    l=len(lst1)

    for i in range(l):

        if lst1[i][0]==en:
            lst1.remove(lst1[i])
            break

    with open("Employee.dat", "wb+") as d:
        for rec in lst1:
            dump(rec,d)

def sal_empno():
    en=int(input("Enter Employee's Number to update the salary  "))
    new_sal=float(input("Enter Employee's New Salary  "))

    lst1=[]

    with open("Employee.dat","rb") as d:
        try:
            while True:
                lst=load(d)
                lst1.append(lst)
        except EOFError:
            print()
    l=len(lst1)

    for i in range(l):
        if lst1[i][0]==en:
            lst1[i][2]=new_sal
            break

    with open("Employee.dat", "wb+") as d:
        for rec in lst1:
            dump(rec,d)

print(" ")
print(" ")
print("----------------------Output-------------------------")
print(" ")
print(" ")

print("Creating File ")
create_file()
menu='''

1) Display all records
2) Append a new employee
3) Search for an employee with employee number
4) Search & display for all occurances of given employee name
5) Sort employee records according to their employee number
6) Delete an employee whose employee number is given
7) Update the salary of an employee whose employee number is given
8)Exit

'''
print(menu)
while True:

    ch=int(input("Enter your choice:  "))

    if ch==1:
        display_file()

    elif ch==2:
        append_file()

    elif ch==3:
        find_empno()
    
    elif ch==4:
        find_ename()

    elif ch==5:
        sort_empno()

    elif ch==6:
        del_empno()

    elif ch==7:
        sal_empno()

    elif ch==8:
        print("End of the program ")
        break
    
    else:
        print("Wrong Input")
```
